```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

### Training Dataset (iris-train)

```python
data = pd.read_excel("iris-train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
```

```python
x
```

```python
y = data.iloc[:, -1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[5.4,	3,	4.5,	1.5]])
```

```python
predictions
```

### Testing dataset (iris-test)

```python
datatest = pd.read_excel("iris-test.xlsx")
```

```python
x_test = datatest.iloc[:, :-1]
```

```python
predictions = model.predict(x_test.values)
```

```python
predictions
```

### Measuring the Accuracy of the Model

```python
actual = datatest.iloc[:,-1]
```

```python
count = 0
for i in range(0,len(actual)):
    if actual[i] == predictions[i]:
        count = count + 1
```

```python
accuracy = (count*100)/len(actual)
```

```python
f"The acuuracy of the dataset is {accuracy}%"
```
